```python
%matplotlib inline
import pylab as pl
pl.style.use('dark_background')
pl.rcParams['figure.figsize'] = (12,10)
```

```python
from astropy import coordinates, units as u
```

```python
from astropy.coordinates import get_body_barycentric, get_body, get_moon
from astroplan import FixedTarget, Observer
```

```python
from astropy.time import Time
```

```python
def eastern(time):
    est = time.to_datetime(timezone=CTO.timezone)
    
    return est.strftime('%H:%M:%S')
```

```python
CTO = Observer(location=coordinates.EarthLocation(lat=29.643018, lon=-82.349004*u.deg, height=31*u.m),
               timezone='EST',
               name='University of Florida Campus Teaching Observatory',
              )
```

```python
time = Time("2021-02-08")
```

```python
solarsystembodies = ['Mars', 'Venus', 'Jupiter', 'Saturn', 'Uranus']
solarsystemtargets = [FixedTarget(get_body(name, time), name=name) for name in solarsystembodies]
```

```python
from astroplan.plots import plot_airmass 
import matplotlib.pyplot as plt 

plot_airmass(solarsystemtargets,
             observer=CTO,
             time=time,
             brightness_shading=True)

plt.legend(loc='best')
plt.show() 
```
